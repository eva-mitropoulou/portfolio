## 1. Goal

Build a baseline EGFR pIC50 workflow from ChEMBL-derived records and compare random split performance with Bemis-Murcko scaffold split performance.

The full project includes ChEMBL EGFR IC50 curation, pIC50 transformation, RDKit descriptors, Morgan fingerprints, and cached metric CSVs. This notebook reads only the public-safe summary JSON.

## 3. Cleaning / Preprocessing

The project filters model-ready activity records, computes descriptors, and stores cached outputs. Raw rows are intentionally not displayed in this notebook.

## 4. Features / Descriptors

## 5. Validation

The key validation comparison is random split versus Bemis-Murcko scaffold split. The scaffold split result is the stronger public claim because it better tests scaffold generalization.

```python
drop = {
    "rmse_increase": round(summary["scaffold_split_morgan_random_forest"]["rmse"] - summary["random_split_morgan_random_forest"]["rmse"], 3),
    "r2_drop": round(summary["random_split_morgan_random_forest"]["r2"] - summary["scaffold_split_morgan_random_forest"]["r2"], 3),
}
drop
```

## 6. Results

Morgan random forest performance dropped from RMSE 0.712 / R2 0.719 under random split to RMSE 0.871 / R2 0.550 under scaffold split. This supports leakage-aware validation framing.
